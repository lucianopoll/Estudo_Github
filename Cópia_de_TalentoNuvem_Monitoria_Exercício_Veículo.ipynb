```python
def classificar_veiculo(quantidade_rodas_veiculo, peso_bruto_veiculo, quantidade_pessoas_veiculo):
  if qtd_rodas == 2 or qtd_rodas == 3:
    return 'A'
  elif peso_bruto_veiculo < 3500 and quantidade_pessoas_veiculo <= 8:
    return 'B'
  elif quantidade_rodas_veiculo >= 4 and peso_bruto >= 3500 and \
       peso_bruto < 6000:
    return 'C'
  else:
    return 'N'
```

```python
# qtd_rodas, peso_bruto, qtd_pessoas = [2, 3200, 10]
# qtd_rodas, peso_bruto, qtd_pessoas = [2, 500, 2]
qtd_rodas, peso_bruto, qtd_pessoas = [4, 3800, 2]
print(qtd_rodas)
print(peso_bruto)
print(qtd_pessoas)
```

```python
resultado = classificar_veiculo(quantidade_rodas_veiculo = qtd_rodas,
                                peso_bruto_veiculo = peso_bruto,
                                quantidade_pessoas_veiculo = qtd_pessoas)
print(resultado)
```

```python
lista_testes = [
    [2, 500, 2]
    ,[2, 3200, 10]
    ,[4, 3800, 2]
    ,[4, 3200, 10]
]
```

```python
for linha_teste in lista_testes:
  qtd_rodas, peso_bruto, qtd_pessoas = linha_teste
  resultado = classificar_veiculo(quantidade_rodas_veiculo = qtd_rodas,
                                peso_bruto_veiculo = peso_bruto,
                                quantidade_pessoas_veiculo = qtd_pessoas)
  print(linha_teste)
  print(resultado)
  print()
```
